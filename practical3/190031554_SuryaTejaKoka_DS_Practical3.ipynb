```python
# using Select -- by position

df.iloc[0,:]
```

```python
# using Select -- by index
df.loc[0]
```

```python
#  Mass renaming of columns
df.rename(columns=lambda x:x+'1')
```

```python
# Selective Renaming
df.rename(columns={'total_rooms' : 'Total_Rooms'})
```

```python
# Change the index
df.set_index('longitude')
```

```python
#  Mass renaming of index

df.rename(index=lambda x:x+1)
```

```python
'''IV.Filter'''

df[df['housing_median_age']>30]
```

```python
'''Sort'''
surya1 = df.sort_values('longitude')
surya1
```

```python
'''Groupby
    Returns a groupby object for values from one column
'''
df.groupby('latitude')
```

```python
# Returns groupby object for values from multiple columns
df.groupby(['latitude','total_rooms'])
```

```python
#  Create a pivot table that groups by col1 and calculates the mean of col2 and col3
df.pivot_table(index='total_rooms',values= ['total_bedrooms','population'],aggfunc=np.mean)
```
